```
# Simple Snake Game in Python 3 for Beginners
# By Ninja-Pythonist

import turtle
import time
import random

delay = 0.1

# Inicia los puntajes
score = 0
high_score = 0

# Abre una ventana para correr el juego.
wn = turtle.Screen()
wn.title("Snake Game by @Ninja-Pythonist")
wn.bgcolor("green")
wn.setup(width=600, height=600)
wn.tracer(0) # Turns off the screen updates

# Cabeza de la serpiente.
head = turtle.Turtle()
head.speed(0)
head.shape("square") #da la forma
head.color("black") #da el color
head.penup()
head.goto(0,0)
head.direction = "stop"

# Alimento de la serpiente.
food = turtle.Turtle()
food.speed(0)
food.shape("circle") #forma
food.color("red") #color
food.penup()
food.goto(0,100)

segments = []

# Tablero de puntos
pen = turtle.Turtle()
pen.speed(0)
pen.shape("square")
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0, 260)
pen.write("Puntaje: 0  Record: 0", align="center", font=("Courier", 24, "normal"))

# Funciones para definir los movimientos.
def go_up():
    if head.direction != "down":
        head.direction = "up"

def go_down():
    if head.direction != "up":
        head.direction = "down"

def go_left():
    if head.direction != "right":
        head.direction = "left"

def go_right():
    if head.direction != "left":
        head.direction = "right"

def move():
    if head.direction == "up":
        y = head.ycor()
        head.sety(y + 20)

    if head.direction == "down":
        y = head.ycor()
        head.sety(y - 20)

    if head.direction == "left":
        x = head.xcor()
        head.setx(x - 20)

    if head.direction == "right":
        x = head.xcor()
        head.setx(x + 20)

# Teclas para mover la serpiente.
wn.listen()
wn.onkeypress(go_up, "w")
wn.onkeypress(go_down, "s")
wn.onkeypress(go_left, "a")
wn.onkeypress(go_right, "d")

# Main del juego.
while True:
    wn.update()

    # Revisa que no choque con los bordes.
    if head.xcor()>290 or head.xcor()<-290 or head.ycor()>290 or head.ycor()<-290:
        time.sleep(1)
        head.goto(0,0)
        head.direction = "stop"

        # Elimina la cola de la serpiente.
        # Reinicia la forma de la serpiente.
        for segment in segments:
            segment.goto(1000, 1000)
        
        # Clear the segments list
        segments.clear()

        # Reinicia Puntaje
        score = 0

        # Reinicia la velocidad del juego
        delay = 0.1

        pen.clear()
        pen.write("Score: {}  High Score: {}".format(score, high_score), align="center", font=("Courier", 24, "normal")) 


    # Revisa si la serpiente pasa por donde hay comida
    if head.distance(food) < 20:
        # Mueve la comida de forma aleatoria
        x = random.randint(-290, 290)
        y = random.randint(-290, 290)
        food.goto(x,y)

        # Agrega cola a la serpiente
        new_segment = turtle.Turtle()
        new_segment.speed(0)
        new_segment.shape("square")
        new_segment.color("grey")
        new_segment.penup()
        segments.append(new_segment)

        # Hace mas rapido el juego
        delay -= 0.001

        # Aumenta el puntaje
        score += 10

        if score > high_score:
            high_score = score
        
        pen.clear()
        pen.write("Score: {}  High Score: {}".format(score, high_score), align="center", font=("Courier", 24, "normal")) 

    # Move the end segments first in reverse order
    for index in range(len(segments)-1, 0, -1):
        x = segments[index-1].xcor()
        y = segments[index-1].ycor()
        segments[index].goto(x, y)

    # Move segment 0 to where the head is
    if len(segments) > 0:
        x = head.xcor()
        y = head.ycor()
        segments[0].goto(x,y)

    move()    

    # Check for head collision with the body segments
    for segment in segments:
        if segment.distance(head) < 20:
            time.sleep(1)
            head.goto(0,0)
            head.direction = "stop"
        
            # Hide the segments
            for segment in segments:
                segment.goto(1000, 1000)
        
            # Clear the segments list
            segments.clear()

            # Reset the score
            score = 0

            # Reset the delay
            delay = 0.1
        
            # Update the score display
            pen.clear()
            pen.write("Score: {}  High Score: {}".format(score, high_score), align="center", font=("Courier", 24, "normal"))

    time.sleep(delay)

wn.mainloop()
```
